```python
import os
import sys

# --- 1. Setup Environment and Paths ---
def setup_environment():
    print("--- Setting up environment and paths ---")
    current_dir = os.getcwd()
    resources_folder = os.path.join(current_dir, '..', 'resources')
    sys.path.append(resources_folder)
    print(f"Current directory: {current_dir}")
    print(f"Resources folder added to path: {resources_folder}")
    print(f"Files in current directory: {os.listdir()}")
    print("-" * 50)

# --- 2. Import Libraries ---
def import_libraries():
    print("--- Importing necessary libraries ---")
    global pd, np, load_dotenv, roc_auc_score, average_precision_score, sc, mt
    import pandas as pd
    import numpy as np
    from dotenv import load_dotenv
    from sklearn.metrics import roc_auc_score, average_precision_score
    import snowflake_connector as sc
    import model_training as mt
    print("Libraries imported successfully.")
    print("-" * 50)

# --- 3. Establish Snowflake Connection ---
def connect_to_snowflake(user):
    print("--- Establishing Snowflake connection ---")
    global sf_conn
    sf_conn = sc.snowflake_conn(user=user)
    print("Snowflake connection established.")
    print("-" * 50)

# --- 4. Fetch Data from Snowflake ---
def fetch_data(query):
    print("--- Fetching data from Snowflake ---")
    global df
    cur = sf_conn.cursor()
    cur.execute(query)
    df = pd.DataFrame.from_records(iter(cur), columns=[x[0] for x in cur.description])
    print(f"Data fetched successfully. Shape: {df.shape}")
    print("-" * 50)

# --- 5. Split Data into Samples ---
def split_data():
    print("--- Splitting data into development, test, and scoring samples ---")
    global data_dev, filters_applied_dev, data_test, filters_applied_test, data_scoring, filters_applied_scoring
    data_dev, filters_applied_dev = mt.build_model_development_sample(data=df)
    data_test, filters_applied_test = mt.build_model_application_sample(data=df)
    data_scoring, filters_applied_scoring = mt.build_model_scoring_sample(data=df)
    print("Data splitting complete.")
    print("-" * 50)

# --- 6. Display Sample Information ---
def display_sample_information():
    print("--- Displaying sample information ---")
    print(f"Original data shape: {df.shape}")
    print(f"Filters applied for model development sample: {filters_applied_dev}")
    print(f"Filters applied for model application sample: {filters_applied_test}")
    print(f"Filters applied for model scoring sample: {filters_applied_scoring}")
    mt.filters_applied_chart(filters_applied_dev, filters_applied_test, filters_applied_scoring)
    print("-" * 50)

# --- 7. Analyze Data Samples ---
def analyze_data_samples():
    print("--- Analyzing data samples ---")
    print(f"Model development Through the door number of transactions: {data_dev.shape[0]}")
    mt.booked_rejected_pie_chart(data_dev)
    mt.default_pie_chart(data_dev)

    print(f"Model application  Through the door number of transactions: {data_test.shape[0]}")
    mt.booked_rejected_pie_chart(data_test)
    mt.default_pie_chart(data_test)

    print(f"Model scoring Through the door number of transactions: {data_scoring.shape[0]}")
    mt.booked_rejected_pie_chart(data_scoring)
    mt.default_pie_chart(data_scoring)
    print("-" * 50)

# --- 8. Define and Apply Risk Grade Path ---
def define_and_apply_risk_grade_path():
    print("--- Defining and applying risk grade path ---")
    global conditions, choices
    conditions = [
        (lambda df: (df['pg_required_c'] == True) & (df['sbfeaccountcount_ln'] >= 1) & (df['sbfehitindex_ln'] >= 2) & (df['fico_score'].notnull())),
        (lambda df: (df['pg_required_c'] == True) & (df['sbfeaccountcount_ln'] >= 1) & (df['sbfehitindex_ln'] >= 2) & (df['fico_score'].isnull())),
        (lambda df: (df['pg_required_c'] == True) & (df['b2bcnt2y_ln'] >= 1) & (df['sbfehitindex_ln'] == 1) & (df['fico_score'].notnull())),
        (lambda df: (df['pg_required_c'] == True) & (df['b2bcnt2y_ln'] >= 1) & (df['sbfehitindex_ln'] == 1) & (df['fico_score'].isnull())),
        (lambda df: (df['pg_required_c'] == True) & ((df['sbfeaccountcount_ln'] < 1) | (df['sbfeaccountcount_ln'].isnull())) & ((df['b2bcnt2y_ln'] < 1) | (df['b2bcnt2y_ln'].isnull())) & ((df['sbfehitindex_ln'] <= 1) | (df['sbfehitindex_ln'].isnull())) & (df['fico_score'].notnull())),
        (lambda df: (df['pg_required_c'] == True) & ((df['sbfeaccountcount_ln'] < 1) | (df['sbfeaccountcount_ln'].isnull())) & ((df['b2bcnt2y_ln'] < 1) | (df['b2bcnt2y_ln'].isnull())) & ((df['sbfehitindex_ln'] <= 1) | (df['sbfehitindex_ln'].isnull())) & (df['fico_score'].isnull())),
        (lambda df: (df['pg_required_c'] == False) & (df['sbfeaccountcount_ln'] >= 1) & (df['sbfehitindex_ln'] >= 2)),
        (lambda df: (df['pg_required_c'] == False) & (df['b2bcnt2y_ln'] >= 1) & (df['sbfehitindex_ln'] == 1)),
        (lambda df: (df['pg_required_c'] == False) & ((df['sbfeaccountcount_ln'] < 1) | (df['sbfeaccountcount_ln'].isnull())) & ((df['b2bcnt2y_ln'] < 1) | (df['b2bcnt2y_ln'].isnull())) & ((df['sbfehitindex_ln'] < 1) | (df['sbfehitindex_ln'].isnull()))),
        (lambda df: (df['pg_required_c'] == False) & ((df['b2bcnt2y_ln'] < 1) | (df['b2bcnt2y_ln'].isnull())) & (df['sbfehitindex_ln'] == 1) & (df['fico_score'].notnull())),
        (lambda df: (df['pg_required_c'] == False) & ((df['b2bcnt2y_ln'] < 1) | (df['b2bcnt2y_ln'].isnull())) & (df['sbfehitindex_ln'] == 1) & (df['fico_score'].isnull()))
    ]

    choices = [
        'pg_and_1_plus_sbfe_tradeline_and_fico_hit',
        'pg_and_1_plus_sbfe_tradeline_and_fico_no_hit',
        'pg_and_no_sbfe_tradeline_and_1_plus_sba_tradeline_and_fico_hit',
        'pg_and_no_sbfe_tradeline_and_1_plus_sba_tradeline_and_fico_no_hit',
        'pg_and_no_sbfe_tradeline_and_no_sba_tradeline_and_fico_hit',
        'pg_and_no_sbfe_tradeline_and_no_sba_tradeline_and_fico_no_hit',
        'no_pg_and_1_plus_sbfe_tradeline',
        'no_pg_and_no_1_plus_sbfe_tradeline_and_1_plus_sba_tradeline',
        'no_pg_no_sbfe_no_sba_no_fico',
        'no_pg_and_no_sbfe_tradeline_and_false_1_plus_sba_tradeline_and_fico_hit',
        'no_pg_and_no_sbfe_tradeline_and_false_1_plus_sba_tradeline_and_fico_no_hit'
    ]

    global data_dev, data_test, data_scoring
    data_dev = mt.apply_risk_grade_path(data_dev, conditions, choices)
    data_test = mt.apply_risk_grade_path(data_test, conditions, choices)
    data_scoring = mt.apply_risk_grade_path(data_scoring, conditions, choices)
    print("Risk grade path applied to all data samples.")
    print("-" * 50)

# --- 9. Define Risk Grade Segments ---
def define_risk_grade_segments():
    print("--- Defining risk grade segments ---")
    global RISK_GRADE_SEGMENTS
    CONDITION_PG_SBFE_FICO = {"column": "risk_grade_path", "allowed_values": ["pg_and_1_plus_sbfe_tradeline_and_fico_hit"]}
    CONDITION_PG_SBFE_NO_FICO = {"column": "risk_grade_path", "allowed_values": ["pg_and_1_plus_sbfe_tradeline_and_fico_no_hit"]}
    CONDITION_PG_SBA_FICO = {"column": "risk_grade_path", "allowed_values": ["pg_and_no_sbfe_tradeline_and_1_plus_sba_tradeline_and_fico_hit"]}
    CONDITION_PG_SBA_NO_FICO = {"column": "risk_grade_path", "allowed_values": ["pg_and_no_sbfe_tradeline_and_1_plus_sba_tradeline_and_fico_no_hit"]}
    CONDITION_PG_FICO_ONLY = {"column": "risk_grade_path", "allowed_values": ["pg_and_no_sbfe_tradeline_and_no_sba_tradeline_and_fico_hit"]}
    CONDITION_PG_NO_HITS = {"column": "risk_grade_path", "allowed_values": ["pg_and_no_sbfe_tradeline_and_no_sba_tradeline_and_fico_no_hit"]}
    CONDITION_NO_PG_SBFE = {"column": "risk_grade_path", "allowed_values": ["no_pg_and_1_plus_sbfe_tradeline"]}
    CONDITION_NO_PG_SBA = {"column": "risk_grade_path", "allowed_values": ["no_pg_and_no_1_plus_sbfe_tradeline_and_1_plus_sba_tradeline"]}
    CONDITION_NO_PG_NO_HITS = {"column": "risk_grade_path", "allowed_values": ["no_pg_no_sbfe_no_sba_no_fico"]}
    CONDITION_NO_PG_FALSE_SBA_FICO_HIT = {"column": "risk_grade_path", "allowed_values": ["no_pg_and_no_sbfe_tradeline_and_false_1_plus_sba_tradeline_and_fico_hit"]}
    CONDITION_NO_PG_FALSE_SBA_NO_HITS = {"column": "risk_grade_path", "allowed_values": ["no_pg_and_no_sbfe_tradeline_and_false_1_plus_sba_tradeline_and_fico_no_hit"]}

    RISK_GRADE_SEGMENTS = {
        "pg_sbfe_ln_and_fico": {"condition": CONDITION_PG_SBFE_FICO, "model_cols": ["intercept", "ln_score", "fico_score_filled"], "train_filter": "booked", "core_model_vars": ["fico_score", "ln_score"]},
        "pg_sbfe_ln_only": {"condition": CONDITION_PG_SBFE_NO_FICO, "model_cols": ["intercept", "ln_score"], "train_filter": "booked", "core_model_vars": ["ln_score"]},
        "pg_sba_ln_and_fico": {"condition": CONDITION_PG_SBA_FICO, "model_cols": ["intercept", "ln_score", "fico_score_filled"], "train_filter": "booked", "core_model_vars": ["fico_score", "ln_score"]},
        "pg_sba_ln_only": {"condition": CONDITION_PG_SBA_NO_FICO, "model_cols": ["intercept", "ln_score"], "train_filter": "booked", "core_model_vars": ["ln_score"]},
        "pg_fico_only": {"condition": CONDITION_PG_FICO_ONLY, "model_cols": ["intercept", "fico_score"], "train_filter": "booked", "core_model_vars": ["fico_score"]},
        "pg_no_hits": {"condition": CONDITION_PG_NO_HITS, "model_cols": ["intercept"], "train_filter": "booked", "core_model_vars": []},
        "no_pg_sbfe_ln_only": {"condition": CONDITION_NO_PG_SBFE, "model_cols": ["intercept", "ln_score"], "train_filter": "booked", "core_model_vars": ["ln_score"]},
        "no_pg_sba_ln_only": {"condition": CONDITION_NO_PG_SBA, "model_cols": ["intercept", "ln_score"], "train_filter": "booked", "core_model_vars": ["ln_score"]},
        "no_pg_no_hits": {"condition": CONDITION_NO_PG_NO_HITS, "model_cols": ["intercept"], "train_filter": "booked", "core_model_vars": []},
        "no_pg_false_sba_fico_hit": {"condition": CONDITION_NO_PG_FALSE_SBA_FICO_HIT, "model_cols": ["intercept"], "train_filter": "booked", "core_model_vars": []},
        "no_pg_false_sba_no_hits": {"condition": CONDITION_NO_PG_FALSE_SBA_NO_HITS, "model_cols": ["intercept"], "train_filter": "booked", "core_model_vars": []},
    }
    print("Risk grade segments defined.")
    print("-" * 50)

# --- 10. Train Models for Each Segment ---
def train_models():
    print("--- Training models for each risk grade segment ---")
    global NORMALIZE_SCORES, data_dev_filtered, models_dict, normalization_dict
    NORMALIZE_SCORES = True
    data_dev_filtered = mt.risk_grade_estimate_preprocessing(data_dev)
    models_dict, normalization_dict = mt.fit_models_for_segments(data_dev_filtered, RISK_GRADE_SEGMENTS, NORMALIZE_SCORES, 'hard_cutoff_augmentation', 10, None)
    print("Model training complete.")
    print("-" * 50)

# --- 11. Display Model Parameters ---
def display_model_parameters():
    print("--- Displaying model parameters ---")
    print(mt.generate_model_parameters_df(models_dict=models_dict))
    print("-" * 50)

# --- 12. Preprocess Test and Scoring Data ---
def preprocess_test_and_scoring_data():
    print("--- Preprocessing test and scoring data ---")
    global data_test_filtered, data_scoring_filtered
    data_test_filtered = mt.risk_grade_estimate_preprocessing(data_test)
    data_scoring_filtered = mt.risk_grade_estimate_preprocessing(data_scoring)
    print("Test and scoring data preprocessing complete.")
    print("-" * 50)

# --- 13. Display Risk Grade Path Counts ---
def display_risk_grade_path_counts():
    print("--- Displaying risk grade path counts ---")
    print(f"Development data risk grade path counts:\n{data_dev_filtered.risk_grade_path.value_counts()}")
    print(f"Test data risk grade path counts:\n{data_test_filtered.risk_grade_path.value_counts()}")
    print(f"Scoring data risk grade path counts:\n{data_scoring_filtered.risk_grade_path.value_counts()}")
    print("-" * 50)

# --- 14. Apply Fitted Models to Data ---
def apply_models_to_data():
    print("--- Applying fitted models to test and scoring data ---")
    global data_dev_scored, data_test_scored, data_scoring_scored
    data_dev_scored = mt.apply_fitted_models_to_data(data=data_dev_filtered, models_dict=models_dict, normalization_dict=normalization_dict, RISK_GRADE_SEGMENTS=RISK_GRADE_SEGMENTS, NORMALIZE_SCORES=NORMALIZE_SCORES)
    data_test_scored = mt.apply_fitted_models_to_data(data=data_test_filtered, models_dict=models_dict, normalization_dict=normalization_dict, RISK_GRADE_SEGMENTS=RISK_GRADE_SEGMENTS, NORMALIZE_SCORES=NORMALIZE_SCORES)
    data_scoring_scored = mt.apply_fitted_models_to_data(data=data_scoring_filtered, models_dict=models_dict, normalization_dict=normalization_dict, RISK_GRADE_SEGMENTS=RISK_GRADE_SEGMENTS, NORMALIZE_SCORES=NORMALIZE_SCORES)
    print("Fitted models applied successfully.")
    print("-" * 50)

# --- 15. Define Risk Grade and Approval Thresholds ---
def define_thresholds():
    print("--- Defining risk grade and approval thresholds ---")
    global new_pd_risk_grade_thresholds, approval_thresholds
    new_pd_risk_grade_thresholds = {
        "1a": (0.00, 0.0009), "1b": (0.0009, 0.0022), "1c": (0.0022, 0.0066), "2a": (0.0066, 0.0110),
        "2b": (0.0110, 0.0165), "2c": (0.0165, 0.0248), "3a": (0.0248, 0.0371), "3b": (0.0371, 0.0464),
        "3c": (0.0464, 0.0557), "4a": (0.0557, 0.0835), "4b": (0.0835, 0.1040), "4c": (0.1040, 0.1280),
        "5a": (0.1280, 0.1570), "5b": (0.1570, 0.1930), "5c": (0.1930, 0.2370), "6a": (0.2370, 0.2910),
        "6b": (0.2910, 0.3570), "6c": (0.3570, 0.5000), "7a": (0.5000, 0.8140), "7b": (0.8140, 1.0000),
    }
    approval_thresholds = {
        "1a": True, "1b": True, "1c": True, "2a": True, "2b": True, "2c": True, "3a": True, "3b": True,
        "3c": True, "4a": False, "4b": False, "4c": False, "5a": False, "5b": False, "5c": False,
        "6a": False, "6b": False, "6c": False, "7a": False, "7b": False,
    }
    print("Risk grade and approval thresholds defined.")
    print("-" * 50)

# --- 16. Define and Apply PG Subsegments ---
def define_and_apply_pg_subsegments():
    print("--- Defining and applying PG subsegments ---")
    global PG_SUBSEGMENTS
    PG_SUBSEGMENTS = {
        "pg_should_be_required": {"condition": {"column": "pg_subsegment", "allowed_values": ["pg_should_be_required"]}},
        "pg_should_not_be_required": {"condition": {"column": "pg_subsegment", "allowed_values": ["pg_should_not_be_required"]}}
    }
    global data_dev_scored, data_test_scored, data_scoring_scored
    data_dev_scored['pg_subsegment'] = np.where((data_dev_scored['number_of_trucks_c'] <= 7) | (data_dev_scored['years_in_business_num_c'] <= 3), 'pg_should_be_required', 'pg_should_not_be_required')
    data_test_scored['pg_subsegment'] = np.where((data_test_scored['number_of_trucks_c'] <= 7) | (data_test_scored['years_in_business_num_c'] <= 3), 'pg_should_be_required', 'pg_should_not_be_required')
    data_scoring_scored['pg_subsegment'] = np.where((data_scoring_scored['number_of_trucks_c'] <= 7) | (data_scoring_scored['years_in_business_num_c'] <= 3), 'pg_should_be_required', 'pg_should_not_be_required')
    print("PG subsegments defined and applied.")
    print("-" * 50)

# --- 17. Calculate Segment Statistics ---
def calculate_segment_statistics():
    print("--- Calculating segment statistics with approval thresholds ---")
    print("Development Data Segment Statistics:")
    segment_stats_dev = mt.calculate_segment_statistics_with_approval(data=data_dev_scored, RISK_GRADE_SEGMENTS=RISK_GRADE_SEGMENTS, risk_grade_thresholds=new_pd_risk_grade_thresholds, approval_thresholds=approval_thresholds, PG_SUBSEGMENTS=PG_SUBSEGMENTS)
    print(segment_stats_dev)

    print("\nTest Data Segment Statistics:")
    segment_stats_test = mt.calculate_segment_statistics_with_approval(data=data_test_scored, RISK_GRADE_SEGMENTS=RISK_GRADE_SEGMENTS, risk_grade_thresholds=new_pd_risk_grade_thresholds, approval_thresholds=approval_thresholds, PG_SUBSEGMENTS=PG_SUBSEGMENTS)
    print(segment_stats_test)

    print("\nScoring Data Segment Statistics:")
    segment_stats_scoring = mt.calculate_segment_statistics_with_approval(data=data_scoring_scored, RISK_GRADE_SEGMENTS=RISK_GRADE_SEGMENTS, risk_grade_thresholds=new_pd_risk_grade_thresholds, approval_thresholds=approval_thresholds, PG_SUBSEGMENTS=PG_SUBSEGMENTS)
    print(segment_stats_scoring)
    print("-" * 50)

# --- 18. Calculate AUC and PR-AUC ---
def calculate_auc_prauc():
    print("--- Calculating AUC and PR-AUC for each segment ---")
    df_performance = data_dev_scored.copy()
    df_performance['segment'] = df_performance['risk_grade_path'].replace({'double_no_hit': 'no_hit_unknown', 'UNKNOWN': 'no_hit_unknown'})
    segment_counts = df_performance['segment'].value_counts()
    auc_results = {}
    prauc_results = {}

    for segment in df_performance['segment'].unique():
        segment_data = df_performance[df_performance['segment'] == segment]
        y_true = None
        y_score = None

        if segment == 'pg_and_1_plus_sbfe_tradeline_and_fico_hit':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_pg_sbfe_ln_and_fico']
        elif segment == 'pg_and_1_plus_sbfe_tradeline_and_fico_no_hit':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_pg_sbfe_ln_only']
        elif segment == 'pg_and_no_sbfe_tradeline_and_1_plus_sba_tradeline_and_fico_hit':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_pg_sba_ln_and_fico']
        elif segment == 'pg_and_no_sbfe_tradeline_and_1_plus_sba_tradeline_and_fico_no_hit':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_pg_sba_ln_only']
        elif segment == 'pg_and_no_sbfe_tradeline_and_no_sba_tradeline_and_fico_hit':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_pg_fico_only']
        elif segment == 'pg_and_no_sbfe_tradeline_and_no_sba_tradeline_and_fico_no_hit':
            segment_data = segment_data.dropna(subset=['pd_pg_no_hits'])
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_pg_no_hits']
        elif segment == 'no_pg_and_1_plus_sbfe_tradeline':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_no_pg_sbfe_ln_only']
        elif segment == 'no_pg_and_no_1_plus_sbfe_tradeline_and_1_plus_sba_tradeline':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_no_pg_sba_ln_only']
        elif segment == 'no_pg_no_sbfe_no_sba_no_fico':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_no_pg_no_hits']
        elif segment == 'no_pg_and_no_sbfe_tradeline_and_false_1_plus_sba_tradeline_and_fico_hit':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_no_pg_false_sba_fico_hit']
        elif segment == 'no_pg_and_no_sbfe_tradeline_and_false_1_plus_sba_tradeline_and_fico_no_hit':
            y_true = (segment_data['is_bad']).astype(int)
            y_score = segment_data['pd_no_pg_false_sba_no_hits']

        if y_true is not None and y_score is not None:
            auc = roc_auc_score(y_true, y_score)
            auc_results[segment] = auc
            prauc = average_precision_score(y_true, y_score)
            prauc_results[segment] = prauc

    print("Segment Counts:\n", segment_counts)
    print("\nAUC Results:\n", auc_results)
    print("\nPR-AUC Results:\n", prauc_results)
    print("-" * 50)

# --- 19. Generate Predictions and Risk Grades for Score Ranges ---
def generate_predictions_and_risk_grades():
    print("--- Generating predictions and risk grades for score ranges ---")
    parameters = pd.DataFrame(mt.generate_model_parameters_df(models_dict=models_dict))
    fico_scores = [600, 620, 640, 660, 680, 700, 720, 740, 760, 780, 800, 820, 840, 850]
    ln_scores = [500, 550, 600, 650, 700, 750, 800, 850, 900]

    for segment in parameters['segment'].unique():
        mt.predict_and_assign_risk_grade_for_segment(segment, parameters, fico_scores, ln_scores, normalization_dict, new_pd_risk_grade_thresholds)
    print("-" * 50)

# --- 20. Generate Lift Charts ---
def generate_lift_charts():
    print("--- Generating lift charts ---")
    # Example lift charts (rest of the lift chart generation code)
    mt.generate_all_lift_charts(data_test_scored, data_scoring_scored, new_pd_risk_grade_thresholds)
    print("-" * 50)

# --- 21. Analyze Booked Rates by Score Bands ---
def analyze_booked_rates():
    print("--- Analyzing booked rates by score bands ---")
    mt.analyze_booked_rates_for_segments(data_dev_scored)
    print("-" * 50)

# --- Main Pipeline Execution ---
if __name__ == "__main__":
    setup_environment()
    import_libraries()
    connect_to_snowflake(user="W515059")
    fetch_data(query='''
        SELECT *
        FROM "SRC_RISK_FRAUD_SUB"."RISK_FRAUD"."OTR_ADJ_RISK_PREPROCESSED_SAMPLE_MODEL_MONITORING";
    ''')
    split_data()
    display_sample_information()
    analyze_data_samples()
    define_and_apply_risk_grade_path()
    define_risk_grade_segments()
    train_models()
    display_model_parameters()
    preprocess_test_and_scoring_data()
    display_risk_grade_path_counts()
    apply_models_to_data()
    define_thresholds()
    define_and_apply_pg_subsegments()
    calculate_segment_statistics()
    calculate_auc_prauc()
    generate_predictions_and_risk_grades()
    generate_lift_charts()
    analyze_booked_rates()
    print("--- Pipeline execution completed! ---")
```
